# 只讀取寄件者電子信箱

```python
pattern=r"From: \".*\" <(.*)>"
x2=re.findall(pattern,contents)
#print(x2)
for x in x2:
    print(x)
```

# 只讀取電子信箱中的寄件機構資訊
ex: james_ngola2002@maktoob.com --> 取maktoob

```python
pattern=r"From: \".*\" <.*@(.*)\..*>"
x2=re.findall(pattern,contents)
#print(x2)
for x in x2:
    print(x)
```

# 結合上面的配對方式, 將寄件者的帳號與機構訊返回

```python
pattern=r"From: \".*\" <(.*)@(.*)\..*>"
x2=re.findall(pattern,contents)
#print(x2)
for x in x2:
    print(x)
```

# 使用正規表達式對email資料進行處理
這裡我們會使用到python其他的套件協助處理(ex: pandas, email, etc)，這裡我們只需要專注在正規表達式上即可，其他的套件是方便我們整理與處理資料。
讀取與切分Email
讀入的email為一個長字串，利用正規表達式切割讀入的資料成一封一封的email，並將結果以list表示。
輸出: [email_1, email_2, email_3, ....]

```python
import re
import pandas as pd
import email

###讀取文本資料:fradulent_emails.txt###
contents2=""
with open('all_emails.txt', 'r', encoding="utf8", errors='ignore') as f2:
    contents2 = f2.read()
    
emails = re.split(r"From r", contents2, flags=re.M)
emails = emails[1:] #移除第一項的空元素
len(emails) #查看有多少封email
    
###切割讀入的資料成一封一封的email###
###我們可以使用list來儲存每一封email###
###注意！這裡請仔細觀察sample資料，看資料是如何切分不同email###
#<your code>#

#len(emails) #查看有多少封email
```

# 從文本中擷取所有寄件者與收件者的姓名和地址


```python
emails_list = [] #創建空list來儲存所有email資訊

for mail in emails[:20]: #只取前20筆資料 (處理速度比較快)
    emails_dict = dict() #創建空字典儲存資訊
    ###取的寄件者姓名與地址###
    
    
    #Step1: 取的寄件者資訊 (hint: From:)
    #<your code>#
    pattern=r"From: \".*\" <.*@.*\..*>"
    form=re.search(pattern,mail)
    #print(form)
    
    #Step2: 取的姓名與地址 (hint: 要注意有時會有沒取到配對的情況)
    #<your code>#
    send_mail=None
    send_name=None
    
    if form is not None:
        pattern_email=r"<(.*@.*\..*)>"
        pattern_name=r"(\".*\")"
        send_mail=re.search(pattern_email,form.group())
        send_name=re.search(pattern_name,form.group())
        if send_mail is not None:
            emails_dict["sender_email"]=send_mail.group()
        if send_name is not None:
            emails_dict["sender_name"]=send_name.group()
        
        #print(emails_dict["sender_email"])
        #print(emails_dict["sender_name"])
    #emails_dict["sender_email"] = sender_mail.group()
    
    #Step3: 將取得的姓名與地址存入字典中
    #<your code>#
    
        
    
    ###取的收件者姓名與地址###
    #Step1: 取的寄件者資訊 (hint: To:)
    #<your code>#
    
    #Step2: 取的姓名與地址 (hint: 要注意有時會有沒取到配對的情況)
    #<your code>#
        
    #Step3: 將取得的姓名與地址存入字典中
    #<your code>#
    
    pattern=r"To:.*"
    to=re.search(pattern,mail)
    print(to)
    
    #Step2: 取的姓名與地址 (hint: 要注意有時會有沒取到配對的情況)
    #<your code>#
    receiver_mail=None
    receiver_name=None
    
    if to is not None:
        pattern_email=r"\w\S*@.*\b"
        pattern_name=r"(?<=\").*(?=\")"
        receiver_mail=re.search(pattern_email,to.group())
        receiver_name=re.search(pattern_name,to.group())
        if receiver_mail is not None:
            emails_dict["receiver_mail"]=receiver_mail.group()
            print(emails_dict["receiver_mail"])
        if receiver_name is not None:
            emails_dict["receiver_name"]=receiver_name.group()
            print(emails_dict["receiver_name"])
        
        
        
        
        
    ###取得信件日期###
    #Step1: 取得日期資訊 (hint: To:)
    date_info = re.search(r"Date:.*", mail)
    
    #Step2: 取得詳細日期(只需取得DD MMM YYYY)
    if date_info is not None:
        date = re.search(r"\d+\s\w+\s\d+", date_info.group())
    else:
        date = None
        
    #Step3: 將取得的日期資訊存入字典中
    if date is not None:
        emails_dict["date_sent"] = date.group()
    else:
        emails_dict["date_sent"] = date
        
        
    ###取得信件主旨###
    #Step1: 取得主旨資訊 (hint: Subject:)
    subject_info = re.search(r"Subject: .*", mail)
    
    #Step2: 移除不必要文字 (hint: Subject: )
    if subject_info is not None:
        subject = re.sub(r"Subject: ", "", subject_info.group())
    else:
        subject = None
    
    #Step3: 將取得的主旨存入字典中
    emails_dict["subject"] = subject
    
    
    ###取得信件內文###
    #這裡我們使用email package來取出email內文 (可以不需深究，本章節重點在正規表達式)
    try:
        full_email = email.message_from_string(mail)
        body = full_email.get_payload()
        emails_dict["email_body"] = body
    except:
        emails_dict["email_body"] = None
    
    ###將字典加入list###
    emails_list.append(emails_dict)
```
